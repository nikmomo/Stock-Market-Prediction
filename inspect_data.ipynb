# Inspect each dataset in pkl format

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the stock data from pkl files
data_QQQ = pd.read_pickle("data_QQQ_processed.pkl")
# data_TECL = pd.read_pickle("data_TECL.pkl")
# data_SPX = pd.read_pickle("data_SPX.pkl")

print(data_QQQ)
```

```python
# Plot line charts for each column of QQQ data, removing NaN values
columns = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
for column in columns:
    plt.figure(figsize=(10, 6))
    cleaned_data = data_QQQ[column].dropna()  # Drop NaN values before plotting
    plt.plot(cleaned_data.index, cleaned_data, label=column)
    plt.title(f'QQQ {column} Over Time')
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```
